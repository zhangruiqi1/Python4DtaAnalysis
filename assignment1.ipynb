```python

L = list()
for i in range(1,500):
    if i%3==0 and i%7==0:
        L.append(i)
print(L)
```

```python
 S=len(L)
print(S)
print('the mean is {}'.format(sum(L)/(S)))
quartile1 = L[int((S)*0.25)+1-1]
quartile2 = L[int((S)*0.50)+1-1]
quartile3 = L[int((S)*0.75)+1-1]
print('the 25th quartile is {}'.format(quartile1))
print('the 50th quartile is {}'.format(quartile2))
print('the 75th quartile is {}'.format(quartile3))
```

```python
from collections import Counter
with open('ques2.txt', 'r') as f:
    p = f.read()
    words = p.split()
    wordCount = len(words)
    print('The number of words: {}'.format(wordCount))
    average = sum(len(word) for word in words)/wordCount
    print('average: {}'.format(average))
```

```python
output = open('ques2.txt', 'r')
lines = output.readlines()
total = list()
for line in lines:
    words = line.strip().split(' ')
    total += words
#print(total)
#print(len(total))
import statistics
charNum = list()
for word in total:
    charNum.append(len(word))
#print(charNum)
print(statistics.mean(charNum))
print(statistics.median(charNum))
print(statistics.mode(charNum))
```
